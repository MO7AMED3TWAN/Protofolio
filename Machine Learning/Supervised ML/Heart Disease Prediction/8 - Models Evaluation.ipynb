```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
dataset=pd.read_csv("heart.csv")
dataset.head()
```

```python
dataset=dataset.drop_duplicates()
```

```python
# logistic regression model
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score
from sklearn.preprocessing import StandardScaler
from sklearn import metrics

X = dataset[['age', 'sex','cp','trestbps','chol']].values
y = dataset['target'].values 

logistic_matrices = {"accuracy" : [], "lose" : [], "precision" : [], "recall" : [], "f1-score" : []}
size = .4

while size >= .1:

    X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=size,random_state=0)
    
    sc = StandardScaler()
    X_train = sc.fit_transform(X_train)
    X_test = sc.transform(X_test)  
    
    classifier = LogisticRegression() 
    classifier.fit(X_train,y_train) 
    y_pred = classifier.predict(X_test)


    accuracy = accuracy_score(y_test, y_pred)
    logistic_matrices["accuracy"].append(accuracy * 100)
    
    logistic_matrices["lose"].append((1-accuracy) * 100)
    
    precision = precision_score(y_test, y_pred)
    logistic_matrices["precision"].append(precision * 100)
    
    recall = recall_score(y_test, y_pred)
    logistic_matrices["recall"].append(recall * 100)
    
    f1Score = f1_score(y_test, y_pred)
    logistic_matrices["f1-score"].append(f1Score * 100)
    
    
    size -= .05
    
logistic_matrices
```

```python
# naive bayes model
from sklearn.model_selection import train_test_split
from sklearn.naive_bayes import GaussianNB
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score
from sklearn import metrics
from sklearn.preprocessing import StandardScaler

X = dataset[['age', 'sex','cp','trestbps','chol']].values
y = dataset['target'].values 

naive_matrices = {"accuracy" : [], "lose" : [], "precision" : [], "recall" : [], "f1-score" : []}
size = .4

while size >= .1:

    X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=size,random_state=0)
    
    sc = StandardScaler()
    X_train = sc.fit_transform(X_train)
    X_test = sc.transform(X_test)

    classifier = GaussianNB()
    classifier.fit(X_train, y_train)
    y_pred = classifier.predict(X_test)

    accuracy = accuracy_score(y_test, y_pred)
    naive_matrices["accuracy"].append(accuracy * 100)
    
    naive_matrices["lose"].append((1-accuracy) * 100)
    
    precision = precision_score(y_test, y_pred)
    naive_matrices["precision"].append(precision * 100)
    
    recall = recall_score(y_test, y_pred)
    naive_matrices["recall"].append(recall * 100)
    
    f1Score = f1_score(y_test, y_pred)
    naive_matrices["f1-score"].append(f1Score * 100)
    
    
    size -= .05
    
naive_matrices
```

```python
# linear SVM model
from sklearn.model_selection import train_test_split
from sklearn.svm import SVC
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score
from sklearn import metrics
from sklearn.preprocessing import StandardScaler

X = dataset[['age', 'sex','cp','trestbps','chol']].values
y = dataset['target'].values 

linear_matrices = {"accuracy" : [], "lose" : [], "precision" : [], "recall" : [], "f1-score" : []}
size = .4

while size >= .1:

    X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=size,random_state=0)
    
    sc = StandardScaler()
    X_train = sc.fit_transform(X_train)
    X_test = sc.transform(X_test)

    classifier = SVC(kernel = 'linear', random_state = 0)
    classifier.fit(X_train, y_train)
    y_pred = classifier.predict(X_test)

    accuracy = accuracy_score(y_test, y_pred)
    linear_matrices["accuracy"].append(accuracy * 100)
    
    linear_matrices["lose"].append((1-accuracy) * 100)
    
    precision = precision_score(y_test, y_pred)
    linear_matrices["precision"].append(precision * 100)
    
    recall = recall_score(y_test, y_pred)
    linear_matrices["recall"].append(recall * 100)
    
    f1Score = f1_score(y_test, y_pred)
    linear_matrices["f1-score"].append(f1Score * 100)
    
    
    size -= .05
    
linear_matrices
```

```python
# kernel SVM model
from sklearn.model_selection import train_test_split
from sklearn.svm import SVC
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score
from sklearn import metrics
from sklearn.preprocessing import StandardScaler

X = dataset[['age', 'sex','cp','trestbps','chol']].values
y = dataset['target'].values 

kernel_matrices = {"accuracy" : [], "lose" : [], "precision" : [], "recall" : [], "f1-score" : []}
size = .4

while size >= .1:

    X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=size,random_state=0)
    
    sc = StandardScaler()
    X_train = sc.fit_transform(X_train)
    X_test = sc.transform(X_test)

    classifier = SVC(kernel = 'rbf', random_state = 0)
    classifier.fit(X_train, y_train)
    y_pred = classifier.predict(X_test)

    accuracy = accuracy_score(y_test, y_pred)
    kernel_matrices["accuracy"].append(accuracy * 100)
    
    kernel_matrices["lose"].append((1-accuracy) * 100)
    
    precision = precision_score(y_test, y_pred)
    kernel_matrices["precision"].append(precision * 100)
    
    recall = recall_score(y_test, y_pred)
    kernel_matrices["recall"].append(recall * 100)
    
    f1Score = f1_score(y_test, y_pred)
    kernel_matrices["f1-score"].append(f1Score * 100)
    
    
    size -= .05
    
kernel_matrices
```

```python
# KNN model
from sklearn.model_selection import train_test_split
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score
from sklearn import metrics
from sklearn.preprocessing import StandardScaler

X = dataset[['age', 'sex','cp','trestbps','chol']].values
y = dataset['target'].values 

knn_matrices = {"accuracy" : [], "lose" : [], "precision" : [], "recall" : [], "f1-score" : []}
size = .4

while size >= .1:

    X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=size,random_state=0)
    
    sc = StandardScaler()
    X_train = sc.fit_transform(X_train)
    X_test = sc.transform(X_test)

    classifier = KNeighborsClassifier(n_neighbors = 5, metric = 'minkowski', p = 2)
    classifier.fit(X_train, y_train)
    y_pred = classifier.predict(X_test)

    accuracy = accuracy_score(y_test, y_pred)
    knn_matrices["accuracy"].append(accuracy * 100)
    
    knn_matrices["lose"].append((1-accuracy) * 100)
    
    precision = precision_score(y_test, y_pred)
    knn_matrices["precision"].append(precision * 100)
    
    recall = recall_score(y_test, y_pred)
    knn_matrices["recall"].append(recall * 100)
    
    f1Score = f1_score(y_test, y_pred)
    knn_matrices["f1-score"].append(f1Score * 100)
    
    
    size -= .05
    
knn_matrices
```

```python
# decision tree model
from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score
from sklearn import metrics
from sklearn.preprocessing import StandardScaler

X = dataset[['age', 'sex','cp','trestbps','chol']].values
y = dataset['target'].values 

decision_matrices = {"accuracy" : [], "lose" : [], "precision" : [], "recall" : [], "f1-score" : []}
size = .4

while size >= .1:

    X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=size,random_state=0)
    
    sc = StandardScaler()
    X_train = sc.fit_transform(X_train)
    X_test = sc.transform(X_test)

    classifier = DecisionTreeClassifier(criterion = 'entropy', random_state = 0)
    classifier.fit(X_train, y_train)
    y_pred = classifier.predict(X_test)

    accuracy = accuracy_score(y_test, y_pred)
    decision_matrices["accuracy"].append(accuracy * 100)
    
    decision_matrices["lose"].append((1-accuracy) * 100)
    
    precision = precision_score(y_test, y_pred)
    decision_matrices["precision"].append(precision * 100)
    
    recall = recall_score(y_test, y_pred)
    decision_matrices["recall"].append(recall * 100)
    
    f1Score = f1_score(y_test, y_pred)
    decision_matrices["f1-score"].append(f1Score * 100)
    
    
    size -= .05
    
decision_matrices
```

```python
# random forest model
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score
from sklearn import metrics
from sklearn.preprocessing import StandardScaler

X = dataset[['age', 'sex','cp','trestbps','chol']].values
y = dataset['target'].values 

random_matrices = {"accuracy" : [], "lose" : [], "precision" : [], "recall" : [], "f1-score" : []}
size = .4

while size >= .1:

    X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=size,random_state=0)
    
    sc = StandardScaler()
    X_train = sc.fit_transform(X_train)
    X_test = sc.transform(X_test)

    classifier = RandomForestClassifier(n_estimators = 10, criterion = 'entropy', random_state = 0)
    classifier.fit(X_train, y_train)
    y_pred = classifier.predict(X_test)

    accuracy = accuracy_score(y_test, y_pred)
    random_matrices["accuracy"].append(accuracy * 100)
    
    random_matrices["lose"].append((1-accuracy) * 100)
    
    precision = precision_score(y_test, y_pred)
    random_matrices["precision"].append(precision * 100)
    
    recall = recall_score(y_test, y_pred)
    random_matrices["recall"].append(recall * 100)
    
    f1Score = f1_score(y_test, y_pred)
    random_matrices["f1-score"].append(f1Score * 100)
    
    
    size -= .05
    
random_matrices
```
